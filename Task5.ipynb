```python
# Social Media – Post Engagement Analyzer

def analyze_engagement(likes_list):
    total_likes = 0

    for like in likes_list:
        total_likes += like

    print("Total Likes:", total_likes)

    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


likes = [200, 300, 250, 300]
analyze_engagement(likes)
```

```python
# Agriculture – Rainfall Adequacy Checker

def rainfall_check(data, required):
    total = 0

    for value in data:
        total += value

    average = total / len(data)

    print("Average Rainfall:", average)

    if average >= required:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


rain_data = [60, 70, 80, 75]
rainfall_check(rain_data, 70)
```

```python
# Social Media – Duplicate Account Detection

def check_duplicates(usernames):
    unique_users = set(usernames)

    if len(unique_users) != len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


users = ["user1", "user2", "user1", "user3"]
check_duplicates(users)
```

```python
# Agriculture – Premium Crop Price Filter

def premium_crops(prices):
    premium_list = []

    for price in prices:
        if price > 2000:
            premium_list.append(price)

    print("Premium Crops:", premium_list)


crop_prices = [1500, 2500, 3200, 1800]
premium_crops(crop_prices)
```

```python
# System Monitoring – Application Health Checker

def system_health(errors):
    print("Error Count:", errors)

    if errors == 0:
        print("System Status: Healthy")
    elif errors <= 5:
        print("System Status: Minor Issues")
    else:
        print("System Status: Critical Issues")


system_health(7)
```

```python
# E-Learning – Attendance Eligibility System

def attendance_check(record):
    present = 0

    for day in record:
        if day == "P":
            present += 1

    percentage = (present / len(record)) * 100

    print("Attendance Percentage:", percentage)

    if percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


attendance = ["P", "P", "A", "P", "P"]
attendance_check(attendance)
```

```python
# Smart Electricity Bill Analyzer

def calculate_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # usage classification
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status


units_used = 250

total_bill, usage = calculate_bill(units_used)

print("Total Bill:", total_bill)
print("Usage Level:", usage)
```
