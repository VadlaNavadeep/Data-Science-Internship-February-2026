```python
# Mobile Recharge Validation System

def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: "))

        if amount < 50:
            print("Recharge must be at least ₹50. Try again.")
            continue

        if amount in valid_plans:
            print("Recharge Successful!")
            break
        else:
            print("Invalid plan. Please retry.")

validate_recharge()
```

```python
# Inventory Reorder Alert System

def check_inventory(stock):
    for product, quantity in stock.items():
        if quantity < 15:
            print(product, "- Reorder Alert")
        else:
            print(product, "- Stock OK")


products = {
    "Rice": 10,
    "Wheat": 25,
    "Sugar": 12,
    "Oil": 18
}

check_inventory(products)
```

```python
# Student Result Processing System

def result_status(marks):
    total = 0

    for m in marks:
        total += m

    average = total / len(marks)

    print("Average Marks:", average)

    if average >= 50:
        return "Pass"
    else:
        return "Fail"


marks_list = [60, 55, 70, 40]

status = result_status(marks_list)
print("Result:", status)
```

```python
# Cab Fare Estimator

def calculate_fare(distance, peak):
    fare = 50 + (distance * 12)

    if peak.lower() == "yes":
        fare += fare * 0.25

    return fare


while True:
    km = float(input("Enter distance in km: "))
    peak_hour = input("Is it peak hour? (yes/no): ")

    total_fare = calculate_fare(km, peak_hour)

    print("Estimated Fare: ₹", round(total_fare, 2))

    retry = input("Do you want to calculate again? (yes/no): ")
    if retry.lower() != "yes":
        break
```

```python
# Attendance Eligibility Checker

def check_eligibility(attendance):
    present = 0

    for day in attendance:
        if day == "P":
            present += 1

    percentage = (present / len(attendance)) * 100

    print("Attendance:", percentage, "%")

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"


record = ["P", "P", "A", "P", "P", "A", "P"]

result = check_eligibility(record)
print(result)
```

```python
# Password Strength Checker

def check_password(password):
    has_digit = False
    has_special = False
    special_chars = "@#$"

    if len(password) < 8:
        return "Weak Password"

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in special_chars:
            has_special = True

    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"


user_password = input("Enter password: ")
print(check_password(user_password))
```
